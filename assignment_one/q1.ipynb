Read Image as Grayscale

```python
# read input image
import cv2 as cv
import numpy as np
import sys
import matplotlib.pyplot as plt

img = cv.imread('runway.png')
if img is None:
    print("Could not read input image")
    sys.exit(1)

print(img.shape)
print(type(img[0][0][0]))

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
print(img_gray.shape)
print(type(img_gray[0][0]))

# show img and img_gray in two graphs using matplotlib
plot, axis = plt.subplots(1, 2, figsize=(10, 5))
axis[0].imshow(img)
axis[0].set_title('Original Image')
axis[0].axis('off')

axis[1].imshow(img_gray, cmap='gray')
axis[1].set_title('Grayscale Image')
axis[1].axis('off')
plot.show()
```

Section<br> 
 (a) Gamma correction with $\gamma = 0.5$<br> 
 (b) Gamma correction with $\gamma = 2$<br> 

```python
# function for gamma correction
def gamma_correction(image, gamma):
    # convert to float and normalize value to 0-1
    img_norm = image / 255.0
    
    #apply gamma correction
    img_corrected = np.power(img_norm, gamma)
    img_corrected = img_corrected * 255.0

    return img_corrected.astype(np.uint8)

img_a = gamma_correction(img_gray, gamma=0.5)
img_b = gamma_correction(img_gray, gamma=2)

plot, axis = plt.subplots(1, 2, figsize=(10, 5))

axis[0].imshow(img_a, cmap='gray')
axis[0].set_title('Gamma 0.5')
axis[0].axis('off')

axis[1].imshow(img_b, cmap='gray')
axis[1].set_title('Gamma 2')
axis[1].axis('off')
plot.show()
```

Understanding how Gamma effect an image

```python
input = np.arange(0,256)
gamma_a = gamma_correction(input, gamma=0.5)
gamma_b = gamma_correction(input, gamma=2)

plt.plot(input, label='original')
plt.plot(gamma_a, label='gamma 0.5') # values less than 1 brightne the image
plt.plot(gamma_b, label='gamma 2') # values grater than 1 darkn the image
plt.legend()
plt.show()
```

 (c) Contrast Stretching
 

```python
import cv2 as cv
import numpy as np
import sys
import matplotlib.pyplot as plt

img = cv.imread('runway.png')
if img is None:
    print("Canno't read the image")
    sys.exit(1)

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

def transform(r, r1, r2):
    return np.where(
        r < r1, 0,
        np.where((r >= r1) & (r <= r2), (r - r1)/(r2-r1), 1)
    )

img_normalized = img_gray.astype(np.float32) / 255.0
img_transformed = transform(img_normalized, 0.2, 0.8)

img_display = (img_transformed * 255).astype(np.uint8)

plot, axis = plt.subplots(1, 2, figsize=(10, 5))
axis[0].imshow(img_gray,cmap='gray')
axis[0].axis('off')
axis[0].set_title('Original')

axis[1].imshow(img_display,cmap='gray')
axis[1].axis('off')
axis[1].set_title('Contrast Stretched')

plt.show()
          
```
