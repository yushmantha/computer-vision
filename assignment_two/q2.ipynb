```python
import cv2
import numpy as np

# Camera parameters
f_mm = 8.0
Z_mm = 720.0
pixel_size_um = 2.2

pixel_size_mm = pixel_size_um * 1e-3
scale_mm_per_pixel = (Z_mm / f_mm) * pixel_size_mm

print(f"Scale: {scale_mm_per_pixel:.4f} mm per pixel")

points = []

def click_event(event, x, y, flags, param):
    global points, img_display

    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"Clicked: ({x}, {y})")
        points.append((x, y))

        cv2.circle(img_display, (x, y), 5, (0, 0, 255), -1)

        cv2.imshow("Click two points", img_display)

        if len(points) == 2:
            p1, p2 = points

            # Draw line
            cv2.line(img_display, p1, p2, (255, 0, 0), 2)

            # Distance in pixels
            dist_px = np.linalg.norm(np.array(p1) - np.array(p2))
            dist_mm = dist_px * scale_mm_per_pixel

            print(f"Distance: {dist_px:.2f} pixels")
            print(f"Distance: {dist_mm:.2f} mm")

            # Show result on image
            cv2.putText(img_display,
                        f"{dist_mm:.2f} mm",
                        (p1[0], p1[1] - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.6,
                        (0, 255, 0),
                        2)

            cv2.imshow("Click two points", img_display)
            #reset
            points = []

# --- Load image ---
img = cv2.imread("earrings.jpg")
img_display = img.copy()

cv2.imshow("Click two points", img_display)
cv2.setMouseCallback("Click two points", click_event)

print("Click two points to measure...")

cv2.waitKey(0)
cv2.destroyAllWindows()
```
